This notebook demonstrates the PVT-GDLA decoder with Gated Differential Linear Attention for medical image segmentation.


```python
from __future__ import annotations
import sys, os; sys.path.insert(0, os.path.abspath('.'))
import torch
import numpy as np
import matplotlib.pyplot as plt
from src.data import make_synthetic, prepare_data, H, W
from src.decoders import build_model, PVTGDLA
```

```python
# Build PVT-GDLA model
model = build_model('pvt_gdla', in_channels=3, base_filters=64, depth=4, num_heads=4, out_channels=1)
print(f"PVT-GDLA model: {sum(p.numel() for p in model.parameters()):,} params")
print("Key innovation: Gated Differential Linear Attention computes two kernelized attention paths,")
print("subtracts them to cancel common-mode noise, and injects nonlinearity via a head-specific gate.")
```

```python
# Forward pass on synthetic data
data = make_synthetic(n=10, seed=42)
x = torch.from_numpy(data['images'][:2])
with torch.no_grad():
    out = model(x)
    pred = (torch.sigmoid(out) > 0.5).float()
print(f"Input:  {x.shape}")
print(f"Output: {out.shape}")
print(f"Pred:   {pred.shape}")
```
